### Timestamp
In this column, we use only 'year', so we extract only year data and drop timestamp and comments

```python
df[['Year']] = df['Timestamp'].str.extract(r'(\d{4})')
df = df.drop(columns=['Timestamp','comments'])
df = df.reindex(columns=['Year', 'Age', 'Gender', 'Country', 'state', 'self_employed',
       'family_history', 'treatment', 'work_interfere', 'no_employees',
       'remote_work', 'tech_company', 'benefits', 'care_options',
       'wellness_program', 'seek_help', 'anonymity', 'leave',
       'mental_health_consequence', 'phys_health_consequence',
       'coworkers', 'supervisor', 'mental_health_interview',
       'phys_health_interview', 'mental_vs_physical', 'obs_consequence'])
df['Year'] = df['Year'].astype(int)
```

```python
df
```

### Age

There are some of them that are unreasonable. We will replace those with empty values.

```python
df['Age'].drop_duplicates()
```

```python
df['Age'] = df['Age'].replace([-29, 329, 99999999999, -1726, -1, 8, 5, 11], "", inplace= False)
```

```python
df
```

### Gender
There are mixed data which will be grouped to 3 categories: Male, Female, and LGBTQ

- start with drop_duplicates() for checking lists of sex we have
- create lists for male and female to replace the data
- call drop_duplicates() to check other left

```python
df['Gender'].drop_duplicates()
```

```python
filter_male = ['M', 'Male','male','m', 'Male-ish','maile', 'something kinda male?', 'Mal','Make', 'Guy (-ish) ^_^','Man','msle', 'Mail', 'Malr','ostensibly male, unsure what that really means','Male ']
filter_female = ['Female','female','F','Woman','f','Femake','woman', 'Female ','femail']
df['Gender'] = df['Gender'].replace(filter_male, 'Male', regex=False, inplace= False)
df['Gender'] = df['Gender'].replace(filter_female, 'Female', regex=False, inplace= False)
```

```python
df['Gender'].drop_duplicates()
```

```python
filter_lgbtq = list(df['Gender'].drop_duplicates().values)[2:]
df['Gender'] = df['Gender'].replace(filter_lgbtq, 'LGBTQ', regex=False, inplace= False)
```

```python
df['Gender'].drop_duplicates()
```

### Number of Employee

This column will be catagorized 3 size of business:
- 1-100 = small 
- 101-500 = medium
- 501-1000 = large

```python
df['no_employees'] = df['no_employees'].replace(['6-25','26-100','1-5'], 'Small', inplace= False)
df['no_employees'] = df['no_employees'].replace('100-500', 'Medium', inplace= False)
df['no_employees'] = df['no_employees'].replace(['500-1000','More than 1000'], 'Large', inplace= False)
```

```python
df['no_employees'].drop_duplicates()
```

### Answer Table

```python
answer_id = []
for i in range(df['Year'].count()):
    answer_id.append(i)

df['user_id'] = answer_id
```

```python
df
```

```python
answer_df = pd.melt(df, id_vars=['user_id','Year'], var_name= 'question', value_name= 'answer_text')
```

```python
answer_df
```

```python
def answer_data():
   
    #create answer id to be primary key
    answer_id = []
    for i in range(answer_df['Year'].count()):
        answer_id.append(i)

    answer_df['answer_id'] = answer_id


    #create dict to store the question abbreviation and question_id
    #replace dict to change the question abbreviation to question_id
    x = {}
    for i in range(len(question)):
        x[question[i]] = question_text[question[i]][0]

    answer_df['question'] = answer_df['question'].replace(x, inplace= False) 
    
    
    #fill 'NAN' with empty values
    answer_df['answer_text'] = answer_df['answer_text'].fillna("")
    
    return answer_df
```

```python
answer_data = answer_data()
```

```python
answer_data
```

```python
table_answer = []
for i in range(answer_df['Year'].count()):
    table_answer.append((
    answer_data['answer_id'][i],
    answer_data['user_id'][i],
    answer_data['answer_text'][i],
    answer_data['Year'][i],
    answer_data['question'][i]
    ))
```

```python
table_answer
```
